# fMRI Preprocessing
## fMRIPrep
For complete instructions, please visit the [fMRIPrep Documentation](https://fmriprep.org/en/stable/).

### Preparation Steps
```` {admonition} Applying for a FreeSurfer License
fMRIPrep uses FreeSurfer tools, which require a [License](https://surfer.nmr.mgh.harvard.edu/registration.html) to run.
````

```` {admonition} Getting a Singularity Image
Spartan uses Apptainer (or previously Singularity) to run external applications (such as fMRIPrep), which requires a Singularity Image. `fmriprep-23.2.1.sif` is already available on Spartan in the AND Lab project directory. If a newer version is needed, you can obtain one by running the following commands line by line in the terminal:
```bash
module load GCC/11.3.0
module load Apptainer/1.2.3
export APPTAINER_TMPDIR=/tmp/<YOUR_USERNAME>/tmp
export APPTAINER_CACHEDIR=/tmp/<YOUR_USERNAME>/cache
mkdir -p $APPTAINER_TMPDIR $APPTAINER_CACHEDIR
apptainer build /data/gpfs/projects/punim2033/fmriprep-<VERSION>.sif docker://nipreps/fmriprep:<VERSION>
```
````

```` {admonition} Adding an IntendedFor Field
For fieldmaps, you need to add an `IntendedFor` field to your fieldmap JSON sidecar files. This tells fMRIPrep which functional runs each fieldmap should be applied to. See the explanation of [IntendedFor](https://bids-specification.readthedocs.io/en/stable/modality-specific-files/magnetic-resonance-imaging-data.html#fieldmap-data), or refer to the example below.
````

## Main Commands
```bash
apptainer exec --cleanenv \
    -B /path/to/input_dir:/data \
    -B /path/to/output_dir:/out \
    -B /path/to/work:/work \
    /path/to/fmriprep-<VERSION>.sif \
    fmriprep /data /out participant \
    --participant-label <SUBJECT_ID> \
```

## Command Options
| **Task**                       | **Command Option**              |
|--------------------------------|---------------------------------|
| Resample anat & func images    | `--output-spaces`               |
| Ignore an aspect of the input  | `--ignore`                      |
| Stop on first error            | `--stop-on-first-crash`         |
| Run anatomical workflows only  | `--anat-onlyRun`                |

Please refer to the [Usage Notes](https://fmriprep.org/en/stable/usage.html#the-command-line-interface-of-the-docker-wrapper) for other options.

```` {admonition} Example JSON file
:class: hint

```bash
{
    "Modality": "MR",
    "MagneticFieldStrength": 7,
    "ImagingFrequency": 297.208264,
    "Manufacturer": "Siemens",
    "ManufacturersModelName": "Investigational_Device_7T_Plus",
    "InstitutionName": "University of Melbourne",
    "InstitutionalDepartmentName": "MRI",
    "InstitutionAddress": "Royal Parade 30,Parkville,Victoria,AU,3052",
    "DeviceSerialNumber": "18978",
    "StationName": "AWP18978",
    "BodyPartExamined": "BRAIN",
    "PatientPosition": "HFS",
    "ProcedureStepDescription": "Neuropaths",
    "SoftwareVersions": "syngo MR E12",
    "MRAcquisitionType": "2D",
    "SeriesDescription": "Reverse_1p6iso_cmrr",
    "ProtocolName": "Reverse_1p6iso_cmrr",
    "ScanningSequence": "EP",
    "SequenceVariant": "SK\\SS",
    "ScanOptions": "PFP\\FS\\EXT",
    "SequenceName": "epfid2d1_130",
    "ImageType": [
        "ORIGINAL",
        "PRIMARY",
        "M",
        "MB",
        "ND",
        "MOSAIC"
    ],
    "NonlinearGradientCorrection": false,
    "SeriesNumber": 37,
    "AcquisitionTime": "16:32:27.355000",
    "AcquisitionNumber": 1,
    "ImageComments": "Unaliased MB6/PE4",
    "SliceThickness": 1.6,
    "SpacingBetweenSlices": 1.6,
    "SAR": 0.120115,
    "EchoTime": 0.0222,
    "RepetitionTime": 0.8,
    "FlipAngle": 45,
    "PartialFourier": 0.875,
    "BaseResolution": 130,
    "ShimSetting": [
        28,
        -114,
        1039,
        620,
        -176,
        -264,
        -44,
        -100
    ],
    "TxRefAmp": 230,
    "PhaseResolution": 1,
    "ReceiveCoilName": "1Tx32Rx_Head",
    "CoilString": "A32",
    "PulseSequenceDetails": "%CustomerSeq%\\cmrr_mbep2d_bold",
    "WipMemBlock": "88df6425-5111-46a5-b283-01da0c4bf129||Sequence: R016 ve12u-ap01/R016b_F50 r/8923c0dc; Feb  4 2020 18:55:00 by eja",
    "RefLinesPE": 96,
    "CoilCombinationMethod": "Sum of Squares",
    "ConsistencyInfo": "N4_VE12U_LATEST_20181126",
    "MatrixCoilMode": "GRAPPA",
    "MultibandAccelerationFactor": 6,
    "PercentPhaseFOV": 100,
    "PercentSampling": 100,
    "EchoTrainLength": 57,
    "PhaseEncodingSteps": 114,
    "AcquisitionMatrixPE": 130,
    "ReconMatrixPE": 130,
    "BandwidthPerPixelPhaseEncode": 24.038,
    "ParallelReductionFactorInPlane": 2,
    "EffectiveEchoSpacing": 0.000320006,
    "DerivedVendorReportedEchoSpacing": 0.000640012,
    "TotalReadoutTime": 0.0412808,
    "PixelBandwidth": 1925,
    "DwellTime": 2e-06,
    "PhaseEncodingDirection": "j",
    "SliceTiming": [
        0,
        0.1725,
        0.3425,
        0.515,
        0.6875,
        0.0575,
        0.23,
        0.4,
        0.5725,
        0.7425,
        0.115,
        0.2875,
        0.4575,
        0.63,
        0,
        0.1725,
        0.3425,
        0.515,
        0.6875,
        0.0575,
        0.23,
        0.4,
        0.5725,
        0.7425,
        0.115,
        0.2875,
        0.4575,
        0.63,
        0,
        0.1725,
        0.3425,
        0.515,
        0.6875,
        0.0575,
        0.23,
        0.4,
        0.5725,
        0.7425,
        0.115,
        0.2875,
        0.4575,
        0.63,
        0,
        0.1725,
        0.3425,
        0.515,
        0.6875,
        0.0575,
        0.23,
        0.4,
        0.5725,
        0.7425,
        0.115,
        0.2875,
        0.4575,
        0.63,
        0,
        0.1725,
        0.3425,
        0.515,
        0.6875,
        0.0575,
        0.23,
        0.4,
        0.5725,
        0.7425,
        0.115,
        0.2875,
        0.4575,
        0.63,
        0,
        0.1725,
        0.3425,
        0.515,
        0.6875,
        0.0575,
        0.23,
        0.4,
        0.5725,
        0.7425,
        0.115,
        0.2875,
        0.4575,
        0.63
    ],
    "ImageOrientationPatientDICOM": [
        0.995698,
        0.0864152,
        -0.0334448,
        -0.0906155,
        0.983503,
        -0.15656
    ],
    "IntendedFor": [
        "func/sub-001_task-RNT_bold.nii.gz",
        "func/sub-001_task-RNT_sbref.nii.gz",
        "func/sub-001_task-safety_run-1_bold.nii.gz",
        "func/sub-001_task-safety_run-1_sbref.nii.gz",
        "func/sub-001_task-safety_run-2_bold.nii.gz",
        "func/sub-001_task-safety_run-2_sbref.nii.gz",
        "func/sub-001_task-safety_run-3_bold.nii.gz",
        "func/sub-001_task-safety_run-3_sbref.nii.gz",
        "func/sub-001_task-safety_run-4_bold.nii.gz",
        "func/sub-001_task-safety_run-4_sbref.nii.gz",
        "func/sub-001_task-darkroom_run-1_bold.nii.gz",
        "func/sub-001_task-darkroom_run-1_sbref.nii.gz",
        "func/sub-001_task-darkroom_run-2_bold.nii.gz",
        "func/sub-001_task-darkroom_run-2_sbref.nii.gz",
        "func/sub-001_task-darkroom_run-3_bold.nii.gz",
        "func/sub-001_task-darkroom_run-3_sbref.nii.gz"
    ],
    "ImageOrientationText": "Tra>Cor(-9.1)>Sag(-1.1)",
    "InPlanePhaseEncodingDirectionDICOM": "COL",
    "ConversionSoftware": "dcm2niix",
    "ConversionSoftwareVersion": "v1.0.20230411",
    "Dcm2bidsVersion": "3.2.0"
}
```
````

```` {admonition} Example Script
:class: hint

```bash
#!/bin/bash

#SBATCH --account=punim2033
#SBATCH --ntasks=1
#SBATCH --cpus-per-task=32
#SBATCH --nodes=1
#SBATCH --array=1
#SBATCH --job-name='fmriprep'
#SBATCH --time=0-20:00:00
#SBATCH -o /data/gpfs/projects/punim2033/safety_learning/slurm_output/fmriprep/slurm.%j.%N.out
#SBATCH -e /data/gpfs/projects/punim2033/safety_learning/slurm_output/fmriprep/slurm.%j.%N.err

# load modules
module purge
module load GCC/11.3.0
module load OpenMPI/4.1.4
module load minc-toolkit/1.9.18.2
module load Apptainer/1.2.3

# Suffix for directory naming
suffix="safety_fmri_prep"

# BIDS root directory
share_dir=/data/gpfs/projects/punim2033/safety_learning
bids_dir=${share_dir}/safety_fmri_bids

# extract sub id from array and change format
export sub=${SLURM_ARRAY_TASK_ID}

if [[ ${#sub} == 1 ]]
then
	export sub=0$sub
fi

# Place outside BIDS root directory, this allows rerun with minimal reprocessing
work_dir=${share_dir}/work/${suffix}
mkdir -p ${work_dir}
out_dir=$share_dir/${suffix}
mkdir -p ${out_dir}

# fmriprep
apptainer exec --cleanenv \
	-B ${bids_dir}:/data \
	-B ${out_dir}:/out \
	-B ${work_dir}:/work \
    -B ./license.txt:/license.txt \
	./fmriprep-23.2.1.sif \
	fmriprep /data /out participant \
    --participant-label ${sub} \
    --nthreads 16 \
    --fs-license-file /license.txt \
	-w /work \
	--skip_bids_validation \
	--output-spaces anat func MNI152NLin2009cAsym:res-native \
    --ignore slicetiming

my-job-stats -a -n -s

```
````
