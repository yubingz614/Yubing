# Pushing Questionnaires to Pavlovia
If you haven’t used GitLab or GitHub before, the basic idea is to clone an empty project folder from the server to your local device, add files or make changes on your computer, and then push those changes back to the server.

## Creating a Project
After creating a Pavlovia account, go to [GitLab Pavlovia](https://gitlab.pavlovia.org/dashboard/projects). Click **New Project** in the top right corner and then **Create Blank Project**.
<img src="../_images/Picture1.png" style="width: 90%;">

You should see something like this:
<img src="../_images/Picture2.png" style="width: 95%;">

You can also share the project through settings in **Memebers**.
<img src="../_images/Picture3.png" style="width: 95%;">

``` {important}
Note that either credits or a Pavlovia license is required to run the experiments. Check with your project leader to see if a lab account is available.
```

## Cloning the Project
```` {admonition} Cloning an Empty Folder to the Local Device
1. On the Pavlovia project page, click **Clone** in the top right corner and copy the **Clone with HTTPS** link.
<img src="../_images/Picture4.png" style="width: 90%;">
</br></br>
2. Install [Git](https://git-scm.com/downloads) onto your local device.
</br></br>
3. In the terminal, navigate to the directory where you would like the project folder to be cloned (e.g., your Desktop).</br>
<img src="../_images/Picture5.png" style="width: 60%;">

Then run the following command:
```python
git clone <YOUR_PAVLOVIA_LINK>
```

A new folder named after your project will be created on your computer.
````

```` {admonition} Upload Files to GitLab
1. Copy and paste all relevant files into this folder:
<img src="../_images/Picture6.png" style="width: 90%;">
</br></br>
2. In the terminal, run the following command:

```python
cd /path/to/YOUR_CLONED_FOLDER          # Navigate to the cloned folder
git add .                               # Stage all files
git commit -m "Add experiment files"    # Commit with a message
git push origin master                  # Push all commits to the server
```

3. You should then be able to see the files on the Gitlab project page:
<img src="../_images/Picture8.png" style="width: 90%;">
````

``` {important}
If prompted for a password, this is likely your local device password. This step may vary depending on your system configuration.</br>
<img src="../_images/Picture7.png" style="width: 40%;">
```

## Making Changes
```` {admonition} Via Code
To update your project after making changes to the cloned folder, run the following commands:
```python
git pull                        # Pull latest files from the server
git status                      # Check what has been changed locally
git add .                       # Stage all modified files
git commit -m "xxx modified"    # Commit with a message
git push origin master          # Push all commits to the server
```
````

``` {admonition} Via Webpage
You can also edit files directly or replace them by uploading new ones on the GitLab project page:</br>
<img src="../_images/Picture9.png" style="width: 90%;">
```

```{important}
When using the first method, always pull the latest version from the server before making changes, especially when collaborating with others on the same project.
```

## Running Your Project on Pavlovia
You should also be able to see your project on Pavlovia by clicking **Dashboard** and then **Experiments**:
<img src="../_images/Picture10.png" style="width: 80%;">

You can now pilot or run your experiment by changing the experiment status!
<img src="../_images/Picture11.png" style="width: 95%;">
