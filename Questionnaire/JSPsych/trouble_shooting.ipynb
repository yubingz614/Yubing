# Trouble Shooting

```{admonition} Blank Webpage
:class: warning
If the `.html` file appears blank in your browser, this usually indicates an error in the script. Right-click the page, select **Inspect**, then navigate to the **Console** tab to view the error details.</br>
<img src="../_images/PictureT1.png" style="width: 90%;">
```

```{admonition} Inspecting Answers While Testing
:class: warning
If you would like to inspect answers while testing your script, you can use 
`console.log(<THE_VARIABLE_TO_CHECK>)` in your code. The output will appear in your 
browser under **Inspect** and then **Console**.
<img src="../_images/PictureT2.png" style="width: 90%;">
```
