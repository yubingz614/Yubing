# BIDS
BIDS is a standardized way of storing and organizing MRI data, which could facilitate further data processing.

## Preparation Steps
```` {admonition} Building a Conda Environment on Spartan
In the terminal, go to your user directory (`/home/<YOUR_SPARTAN_USERNAME>`) and run the following codes line by line to create an environment:

```bash
module load Anaconda3/2022.10
conda create --name dcm2bids_env
eval "$(conda shell.bash hook)"
conda activate dcm2bids_env
```
````

```` {admonition} Downloading *dcm2bids* Packages
Then run the following codes to download relevant packages to your environment:
```bash
conda install -c conda-forge dcm2niix
conda install -c conda-forge dcm2bids
```
````

```` {admonition} Creating a Configuration File
*dcm2bids* uses the configuration files to determine which data to convert and how to rename it. Each entry in a config file defines a scan to convert, with the following fields:
1. **datatype**: the BIDS data type (e.g., `anat` for anatomical, `func` for functional, `fmap` for fieldmap).
2. **suffix**: the BIDS suffix identifying the scan type (e.g., `T1w`, `bold`, `sbref`, `epi`).
3. **custom_entities**: additional BIDS entities such as task name or run number (e.g., task-pre, run-1).
4. **criteria**: DICOM metadata fields used to identify and match the correct scan. *dcm2bids* compares these values against the DICOM headers of each scan to determine which files to convert. Commonly used fields are:
    - `SeriesDescription`: the name of the scan sequence as set on the MRI scanner (e.g., `MP2RAGE*` means any series description starting with MP2RAGE).
    - `ImageType`: specifies the type of image data (e.g., `ORIGINAL` vs `DERIVED`, `PRIMARY`, reconstruction method). This is useful when multiple scans share the same `SeriesDescription` but differ in how the image was processed.

See the [Configuration File Setup](https://unfmontreal.github.io/Dcm2Bids/3.2.0/tutorial/first-steps/#setting-up-the-configuration-file) for more details, or refer to the example below.
````

## Main commands
```bash
dcm2bids \
    -d /path/to/input_dir \
    -p <SUBJECT_ID> \
    -c /path/to/config.json \
    -o /path/to/output_dir
```

## Command Options
| **Task**                       | **Command Option**              |
|--------------------------------|---------------------------------|
| Skip DICOM to NIFTI conversion | `--skip_dcm2niix`               |
| Overwrite previous output      | `--force_dcm2bids`              |
| BIDS validation check          | `--bids_validate`               |    

```` {admonition} Example Config File
:class: hint

```bash
{
    "descriptions": [
      {
        "datatype": "anat",
        "suffix": "T1w",
        "criteria":{
          "SeriesDescription": "MP2RAGE*",
          "ImageType": ["DERIVED", "PRIMARY", "M", "UNI", "DIS3D", "DIS2D"]
        }
      },
      {
        "datatype": "func",
        "suffix": "bold",
        "custom_entities": "task-pre",
        "criteria":{
          "SeriesDescription": "Pre_1p6iso_cmrr",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "MB", "ND", "MOSAIC"]
        }
      },
      {
        "datatype": "func",
        "suffix": "sbref",
        "custom_entities": "task-pre",
        "criteria":{
          "SeriesDescription": "Pre_1p6iso_cmrr_SBRef",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "ND", "MOSAIC"]
        }    
      },
      {
        "datatype": "func",
        "suffix": "bold",
        "custom_entities": "task-safety_run-1",
        "criteria":{
          "SeriesDescription": "Safety1_1p6iso_cmrr",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "MB", "ND", "MOSAIC"]
        }
      },
      {
        "datatype": "func",
        "suffix": "sbref",
        "custom_entities": "task-safety_run-1",
        "criteria":{
          "SeriesDescription": "Safety1_1p6iso_cmrr_SBRef",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "ND", "MOSAIC"]
        }    
      },
      {
        "datatype": "func",
        "suffix": "bold",
        "custom_entities": "task-safety_run-2",
        "criteria":{
          "SeriesDescription": "Safety2_1p6iso_cmrr",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "MB", "ND", "MOSAIC"]
        }
      },
      {
        "datatype": "func",
        "suffix": "sbref",
        "custom_entities": "task-safety_run-2",
        "criteria":{
          "SeriesDescription": "Safety2_1p6iso_cmrr_SBRef",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "ND", "MOSAIC"]
        }    
      },
          {
        "datatype": "func",
        "suffix": "bold",
        "custom_entities": "task-safety_run-3",
        "criteria":{
          "SeriesDescription": "Safety3_1p6iso_cmrr",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "MB", "ND", "MOSAIC"]
        }
      },
      {
        "datatype": "func",
        "suffix": "sbref",
        "custom_entities": "task-safety_run-3",
        "criteria":{
          "SeriesDescription": "Safety3_1p6iso_cmrr_SBRef",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "ND", "MOSAIC"]
        }    
      },
          {
        "datatype": "func",
        "suffix": "bold",
        "custom_entities": "task-safety_run-4",
        "criteria":{
          "SeriesDescription": "Safety4_1p6iso_cmrr",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "MB", "ND", "MOSAIC"]
        }
      },
      {
        "datatype": "func",
        "suffix": "sbref",
        "custom_entities": "task-safety_run-4",
        "criteria":{
          "SeriesDescription": "Safety4_1p6iso_cmrr_SBRef",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "ND", "MOSAIC"]
        }    
      },
      {
        "datatype": "func",
        "suffix": "bold",
        "custom_entities": "task-post",
        "criteria":{
          "SeriesDescription": "Post_1p6iso_cmrr",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "MB", "ND", "MOSAIC"]
        }
      },
      {
        "datatype": "func",
        "suffix": "sbref",
        "custom_entities": "task-post",
        "criteria":{
          "SeriesDescription": "Post_1p6iso_cmrr_SBRef",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "ND", "MOSAIC"]
        }    
      },
      {
        "datatype": "func",
        "suffix": "bold",
        "custom_entities": "task-gen_run-1",
        "criteria":{
          "SeriesDescription": "Gen1_1p6iso_cmrr",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "MB", "ND", "MOSAIC"]
        }
      },
      {
        "datatype": "func",
        "suffix": "sbref",
        "custom_entities": "task-gen_run-1",
        "criteria":{
          "SeriesDescription": "Gen1_1p6iso_cmrr_SBRef",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "ND", "MOSAIC"]
        }    
      },
      {
        "datatype": "func",
        "suffix": "bold",
        "custom_entities": "task-gen_run-2",
        "criteria":{
          "SeriesDescription": "Gen2_1p6iso_cmrr",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "MB", "ND", "MOSAIC"]
        }
      },
      {
        "datatype": "func",
        "suffix": "sbref",
        "custom_entities": "task-gen_run-2",
        "criteria":{
          "SeriesDescription": "Gen2_1p6iso_cmrr_SBRef",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "ND", "MOSAIC"]
        }    
      },
      {
        "datatype": "func",
        "suffix": "bold",
        "custom_entities": "task-gen_run-3",
        "criteria":{
          "SeriesDescription": "Gen3_1p6iso_cmrr",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "MB", "ND", "MOSAIC"]
        }
      },    
      {
        "datatype": "func",
        "suffix": "sbref",
        "custom_entities": "task-gen_run-3",
        "criteria":{
          "SeriesDescription": "Gen3_1p6iso_cmrr_SBRef",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "ND", "MOSAIC"]
        }    
      },                    
      {
        "datatype": "fmap",
        "suffix": "epi",
        "custom_entities": "dir-PA",
        "criteria":{
          "SeriesDescription": "Reverse_1p6iso_cmrr",
          "ImageType": ["ORIGINAL", "PRIMARY", "M", "MB", "ND", "MOSAIC"]
        }
      }
    ]
  }

```
````

```` {admonition} Example Script
:class: hint

```bash
#!/bin/bash

#SBATCH --account=punim2033
#SBATCH --ntasks=1
#SBATCH --cpus-per-task=3
#SBATCH --mem=48G
#SBATCH --array=1
#SBATCH --job-name='dcm2bids'
#SBATCH --time=0-02:00:00
#SBATCH --partition=sapphire
#SBATCH -o /data/gpfs/projects/punim2033/neuropaths/slurm_output/dcm2bids/slurm.%j.%N.out
#SBATCH -e /data/gpfs/projects/punim2033/neuropaths/slurm_output/dcm2bids/slurm.%j.%N.err

# get subject id from array
export sub=${SLURM_ARRAY_TASK_ID}

# make sure subject id is 3 digits
if [[ ${#sub} -eq 1 ]]; then
    export sub="00$sub"
elif [[ ${#sub} -eq 2 ]]; then
    export sub="0$sub"
fi

echo "Starting dcm2bids for subject '${sub}'"

id=$((${SLURM_ARRAY_TASK_ID}))
ses=1 # only change if participants have done more than one session

# load relevant modules
module load unimelb-mf-clients

# mflux dir
mflux_dir=/projects/proj-5540_neuropaths-1128.4.1290

# create temporary directories to store data while on Spartan
export tmp_in=/tmp/dcm2nii_in/${sub}
export tmp_out=/tmp/dcm2nii_out
mkdir -p $tmp_in; mkdir -p $tmp_out

unimelb-mf-download \
	--mf.config ~/.Arcitecta/mflux.cfg \
	--csum-check \
	--nb-workers 3 \
	--out $tmp_in \
	$mflux_dir/nifti/1.7.188/1.7.188.${id}/1.7.188.${id}.1/1.7.188.${id}.1.${ses}/nifti_series

# activate conda env
module load Anaconda3/2022.10
eval "$(conda shell.bash hook)"
conda activate dcm2bids_env

bids_dir=${tmp_out}/neuropaths_fmri_bids

# check if there's already a dcm2bids dir in the folder we're trying to make one
if [ ! -d $bids_dir ]
then	
	dcm2bids_scaffold -o $bids_dir
else
	echo "BIDS directory already exists, so proceeding without creating a new one."
fi

find "$tmp_in" -type f -name "*.nii" -exec gzip -f {} \;
find "$tmp_in" -mindepth 2 -type f -exec mv -n -t "$tmp_in" -- {} +

config_file=/data/gpfs/projects/punim2033/neuropaths/scripts/neuropaths_dcm2bids_config.json

# transfering data into BIDS format
dcm2bids -d $tmp_in \
	-p $sub \
	-c $config_file \
	-o $bids_dir \
	--skip_dcm2niix \
	--force_dcm2bids

# send files back to mflux
unimelb-mf-upload \
	--mf.config ~/.Arcitecta/mflux.cfg \
	--csum-check --nb-workers 3 \
	--namespace $mflux_dir/neuropaths_fmri_bids \
	$bids_dir/sub-${sub}

cp -r $bids_dir/sub-${sub} /data/gpfs/projects/punim2033/neuropaths/neuropaths_fmri_bids

my-job-stats -a -n -s

```
````
