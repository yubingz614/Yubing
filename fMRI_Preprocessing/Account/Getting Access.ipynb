# Getting Access
This page will guide you through setting up access to DaRIS, Mediaflux, and Spartan. Approval may take a few days. Please submit your access requests early, so everything is in place when you need to get started!

## DaRIS
For complete instructions, please visit the [DaRIS Guide](https://rcs-knowledge-hub.atlassian.net/wiki/spaces/KB/pages/5474892/DaRIS).

```{admonition} Requesting Access
To request access to an existing DaRIS project, submit a request form via [UoM Staff](https://go.unimelb.edu.au/tu78) or [UoM Students](https://go.unimelb.edu.au/6o78).

In the form, select **Research Data Storage** as the service. In the field “Describe your request”, include:
1.  The **DaRIS project ID** you want to access
2.  The access level required (usually **guest**)
```

```{admonition} Accessing the DaRIS Web Portal
After getting the accounts enabled, you'll be able to access DaRIS via the [Web Portal](https://daris.researchsoftware.unimelb.edu.au/daris/DaRIS.html?allowUnsupportedBrowser=true).

Fill in the following information to login:
1.  Domain: `unimelb` (for staff) or `unimelb-student` (for students)  
2.  User: `<YOUR_UNIMELB_USERNAME>`
3.  Password: `<YOUR_UNIMELB_PASSWORD>`

<img src="../_images/PictureG1.png" style="width: 50%;">
```

## Mediaflux
For complete instructions, please visit the [Mediaflux Guide](https://rcs-knowledge-hub.atlassian.net/wiki/spaces/KB/pages/5472333/Mediaflux).

```{admonition} Requesting Access & New Data Storage
Please contact your project leader to be added to an existing Mediaflux project.

New data storage allocations can be requested by any eligible UoM researcher in the [Research Computing Portal](https://rcp.research.unimelb.edu.au/home/).
```

```{admonition} Setting Up Multifactor Authentication (MFA)
*Skip this step if it has not been made mandatory... This will make things much easier...*

1. Download a smartphone app called Mediaflux Pocket (for [Android](https://rcs-knowledge-hub.atlassian.net/wiki/spaces/KB/pages/21397506/Multifactor+Authentication+MFA#Android-devices) and [Apple](https://rcs-knowledge-hub.atlassian.net/wiki/spaces/KB/pages/21397506/Multifactor+Authentication+MFA#Apple-devices) devices).

2. Go to the [Mediaflux Pocket Registration Portal](https://mediaflux.researchsoftware.unimelb.edu.au/mfp-registration) on your PC and log in with your usual Mediaflux login details.

3. Click the QR code to unblur in the resulting screen. Open the Mediaflux Pocket app on your phone, click the orange **+** button in the bottom right of the app, and scan the QR code.

From now on, each time you log in to Mediaflux, you will need to approve a push notification in the Mediaflux Pocket app on your phone to complete the login. For more details, see [Multifactor Authentication (MFA)](https://rcs-knowledge-hub.atlassian.net/wiki/spaces/KB/pages/21397506/Multifactor+Authentication+MFA).
```

```{admonition} Applying for a Mediaflux Token
Secure identity tokens are an alternative way to authenticate with Mediaflux without using your domain username and password. They are useful for bypassing MFA push notifications when accessing Mediaflux and managing data through Spartan. To apply for a token: 

1. Log in to the [Token Portal](https://mediaflux.researchsoftware.unimelb.edu.au/token-portal/) and then click on the **Tokens** icon.

2. Click on the **Create** button at the top left, set a description for the token, and click **Create**. An email with your token will be sent to your UoM email address.

3. Click on the **Roles** tab for the token you just created, then click **Add**. Select **participant-acmd-n** in the pop-up window.

See [Requesting Tokens](https://rcs-knowledge-hub.atlassian.net/wiki/spaces/KB/pages/511541253/Requesting+tokens) for more details.
```

```{admonition} Accessing Mediaflux
You can download Mediaflux Explorer installers below:
1.  [Mediaflux Explorer 1.6.8 for Windows](https://www.arcitecta.com/software/mexplorer/native/windows/Mediaflux%20Explorer-1.6.8.exe)
2.  [Mediaflux Explorer 1.6.8 for macOS](https://www.arcitecta.com/software/mexplorer/native/mac/Mediaflux%20Explorer-1.6.8-x64.dmg)
3.  [Mediaflux Explorer 1.6.8 for Linux](https://www.arcitecta.com/software/mexplorer/native/linux/mediaflux-explorer-1.6.8.zip)

Or simply use their [Web Portal](https://mediaflux.researchsoftware.unimelb.edu.au/desktop/?dti=no).

<img src="../_images/PictureG2.png" style="width: 90%;">

Fill in the following information to login:
1.  Domain: `unimelb` (for staff) or `unimelb-student` (for students)  
2.  User: `<YOUR_UNIMELB_USERNAME>`
3.  Password: `<YOUR_UNIMELB_PASSWORD>`
```
