# Files and Folders
Below is a brief overview of the files and folders in the Safety Learning Project.

```{admonition} The **.html** File
The `.html` file is the main file that brings everything together. You can open it in a web browser to preview the questionnaire, or in a code editor to make modifications (e.g., [Visual Studio Code](https://code.visualstudio.com/)).
<img src="../_images/PictureI1.png" style="width: 100%;">
```

```{admonition} The **other_blocks** Folder
All individual questionnaires are coded in separate files within the **other_blocks** folder.
```

```{admonition} The **jspsych** Folder
You generally do not need to modify the plugins or any files in this folder. It is best to leave it as is.</br>
```
