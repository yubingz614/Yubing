# Mediaflux
Mediaflux is the University’s research data management platform. It allows only limited data handling through Mediaflux Explorer or the Web Portal (e.g., deleting files). All other operations need to be carried out through Spartan (see procedures below).

## Preparation Steps
```` {admonition} Setting Up a Configuration File
1. Create a folder called `.Arcitecta` at your user directory: `/home/<YOUR_SPARTAN_USERNAME>`
2. Create a file named `mflux.cfg` in the folder
3. Add the following to the file:
```bash
host=mediaflux.researchsoftware.unimelb.edu.au
port=443
transport=https
domain=unimelb (for staff) or student (for students)
user=<YOUR_SPARTAN_USERNAME>
token=<TOKEN>
```
In this way, the unimelb-mf-clients commands know how to connect to Mediaflux without you having to type your credentials and server details every time you run a command.
````

## Main Commands
| **Task**                       | **Command**                     |
|--------------------------------|---------------------------------|
| Upload data to Mediaflux       | `unimelb-mf-upload`             |
| Download data from Mediaflux   | `unimelb-mf-download`           |
| Checking data integrity        | `unimelb-mf-check`              |

## Command Options
| **Task**                       | **Command Option**              |
|--------------------------------|---------------------------------|
| Verification of file integrity | `--csum-check`                  |
| Parallel uploads/downloads     | `--nb-workers`                  |

See [Managing Data](https://dashboard.hpc.unimelb.edu.au/data_management/) for more details.

```` {admonition} Example Script
:class: hint

```bash
#!/bin/bash

#SBATCH --account=punim2033
#SBATCH --ntasks=1
#SBATCH --cpus-per-task=3
#SBATCH --mem=48G
#SBATCH --nodes=1
#SBATCH --job-name='mediaflux'
#SBATCH --time=0-10:00:00
#SBATCH --partition=sapphire
#SBATCH -o /data/gpfs/projects/punim2033/safety_learning/slurm_output/mediaflux/slurm.%j.%N.out
#SBATCH -e /data/gpfs/projects/punim2033/safety_learning/slurm_output/mediaflux/slurm.%j.%N.err

# load module
module load unimelb-mf-clients

# set directories
mflux_dir=/path/to/your/mediaflux_project
spartan_dir=/path/to/your/spartan_project

# download from mflux
unimelb-mf-download \
	--mf.config ~/.Arcitecta/mflux.cfg \
	--csum-check \
	--nb-workers 3 \
	--out $spartan_dir/data \
	$mflux_dir/data

# send data back to mflux
unimelb-mf-upload \
	--mf.config ~/.Arcitecta/mflux.cfg \
	--csum-check \
	--nb-workers 3 \
	--namespace $mflux_dir/data \
	$spartan_dir/data
```
````
