# Introduction to JSPsych
JSPsych is a JavaScript framework for creating behavioral experiments or surveys that run in a web browser.

In this guide, we will focus on surveys and use the [Safety Learning Project](https://github.com/ANDLabUniMelb/Questionnaires/tree/main/Projects/Safety%20Learning) as an example. This project includes a consent form, a demographic survey, and 19 psychological scales.

Since each scale is stored as a separate JSPsych file, you can pick and reuse any questionnaire we've already coded when building your own project. All coded questionnaires are available under [Coded Questionnaire Blocks](https://github.com/ANDLabUniMelb/Questionnaires/tree/main/Coded%20Questionnaire%20Blocks). If a questionnaire hasn't been used in the lab before, you can also code it from scratch (see [Modifying an Individual File](../JSPsych/individual_file.ipynb)).

Later sections will walk you through how to modify questionnaire files and make any additional adjustments.
